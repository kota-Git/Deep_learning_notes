# **How does bilogical Neuron get buildup?**

In the 1950s, research on child brain development revealed fascinating insights. At Harvard University, scientists studied a small piece of a child's brain and made significant observations:

A newborn baby has minimal interconnections between neurons. As shown in the diagram, the neuron edges have very few connections to other neurons. However, as the child grows, particularly up to the age of 6, the brain rapidly develops numerous neural connections. This growth is driven by the child’s interaction with their environment, such as observing objects, learning language, forming sentences, and interacting with family. This process is known as biological learning.
image.png
By the age of 6, some neurons in the child’s brain become thicker, reflecting the accumulation of knowledge and experiences. The more the child learns, the thicker these neural connections become. In mathematical terms, this is described as increased "weights," which indicate thicker and stronger connections. These thicker neurons require more energy and calories to function effectively.

After the age of 14, researchers observed a process called synaptic pruning. During this phase, neurons that are no longer useful are disconnected, while others either become thinner or thicker depending on their relevance and usage.

As people age beyond 80, they often experience difficulties in recognizing others, hearing, and sensory perception. This is because, with age, the body becomes weaker, leading to reduced energy levels. Consequently, some neurons in the brain begin to die. This loss of neurons affects cognitive functions, sensory processing, and overall brain performance, contributing to these challenges.

# Single Layer Perceptron Model (in one word):

One of the easiest ANN types.
It consists of a feed-forward network and includes a threshold transfer inside the model.
The main objective of the single-layer perceptron model is to analyze the linearly separable objects with binary outcomes.
A single-layer perceptron can learn only linearly separable patterns.

# **Multilayer perceptron**

A M.L.P. is a deep A.N.N. It has 3 layers including one hidden layer. If it has more than 1 hidden layer, it is called a deep A.N.N.

The M.L.P. is a typical example of a feed-forward A.N.N.

M.L.P. consists of:
→ Input layer
→ Hidden layer
→ Output layer

# Input layers:

The Input layers receive the input signal to be processed from the outside world to the network and together are referred to as the "Input layers."

# Hidden layers:

Hidden layers perform computation and transfer information from the input layer to the output while transforming data. This data existed in non-linear response.
The data existed in complex patterns to understand.
This means the data existed in complex patterns.
Thus, we use hidden layers. In this layer, we separate data in non-linear ways using an activation function. To separate linear ways, we use A.Fs like sigmoid, Relu, or tanh.

# Output layers:

The output layer combines different ways into one effective pattern for smoothing of data.
For this, we use "softmax" A.F. It is responsible for transforming data into outside worlds.

#**How Multilayer Perceptron Captures Non-Linearity in Data**

To overcome this limitation, the multilayer perceptron (MLP) was introduced. MLP networks contain multiple layers, including hidden layers, which introduce non-linear transformations through activation functions. This allows the network to capture complex non-linear relationships between input features and output classes. By incorporating multiple layers and non-linear activation functions, MLP networks can build decision boundaries that can flexibly adapt to non-linear data patterns

# What is the role of weight and bias in ANN?

# Weights:
Weights are parameters in an ANN that decide how much influence the input will have on the output.

If inputs are not so important, they will be scaled down, and if inputs are important, they will be scaled up.

#Bias:
The role of bias is to control the activation of a node.

Bias can delay the activation; similarly, it can accelerate the activation.
